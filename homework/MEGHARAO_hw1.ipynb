<h2> Problem 1 </h2>

Write a function called sumsquares that takes in an integer returns the sum of the squares from 1 to n. For example, sumsquares(3) should return 14 because 1^2+2^2+3^2=14.

Test cases:
sumsquares(3)=14
sumsquares(1)=1
sumsquares(4)=30

Note: You should use a for loop. The point of this exercise is to practice basic programming.

```python
# Put your solution here (you can add more cells using the Insert menu)
def sumsquares(n):
  sum = 0
  for i in range(1, n+1): # process numbers 1 to n
    sum += i**2 # add together
  return sum
```

```python
# test cases
print(sumsquares(3))
print(sumsquares(1))
print(sumsquares(4))
```

<h2> Problem 2 </h2>

Write a function that takes in an integer returns the sum of all the numbers between 1 and n inclusive that are divisible by 5. For example, f(12) should return 15 because the numbers between 1 and n that are divisible by 5 in between 1 and 12 are 5 and 10.

Testcases:

f(12)=15
f(15)=30
f(16)=30
f(20)=50

```python
# Solution here
def f(n):
  numbers = [i for i in range(1, n+1) if i%5 ==0]
  return(sum(numbers))
```

```python
# test cases
print(f(12))
print(f(15))
print(f(16))
print(f(20))
```

<h2> Problem 3 </h2>

Write a function that takes an input n, and returns the nth fibonacci number. Recall the fibonacci sequence 1 1 2 3 5 8 is generated as follows: the n+1st number is the sum of the n-1st number and the nth number.
For example,
f(3)=2 and f(4)=3.

Testcases:

f(4)=3
f(5)=5

Do not use recursion; use a for loop.

```python
# Solution here
def fibonacci(n):
  f0 = 1
  f1 = 1
  fib_num = 0
  for i in range(n):
    # run through the sequence of fibonacci numbers from 1 to n
    fib_num = f0
    # update fibonacci numbers
    f0 = f1
    f1 = fib_num + f1
  return fib_num
```

```python
# test cases
print(fibonacci(1))
print(fibonacci(2))
print(fibonacci(3))
print(fibonacci(4))
print(fibonacci(5))
print(fibonacci(6))
```

<h2> Problem 4 </h2>
Write a function that determines if a number is prime.
Your function should **only** use for loops, if statements, and the % operator. It should return either True or False.

Testcases:

f(5)=True
f(3)=True
f(1)=False
f(12)=False


```python
def is_prime(n):
  if n == 1:
    return False # 1 is not a prime number
  for i in range (2, int(n**0.5)): # loop through all possible factors of n
    if n%i == 0:
      return False # if anything can divide n evenly, n is not prime
  return True # if no divisble factors, n must be prime
```

```python
# test cases
print(is_prime(51))
print(is_prime(53))
print(is_prime(1))
print(is_prime(12))
```

<h2> Problem 5 </h2>

Write a function that finds the remainder for positive integer division.
For example, remainder(10,3)=1 because 10/3 has remainder 1.
Your function should only use: while loops and subtraction.

The point is to make your own version of the % operator.

Testcases:

f(10,2)=0
f(10,3)=1
f(12,4)=0


```python
def remainder(x, y):
  while x >= y:
    x -= y
  return x
```

```python
# test cases
print(remainder(10,2))
print(remainder(10,3))
print(remainder(12,4))
print(remainder(23,5))
```
