### Problem 2

Write a function freq, given a lower case string, returns the frequency of each letter in the form of a dictionary.
E.g. 
freq('hello')={'h':1,'e':1,'l':2, 'o':1}


```python
def freq(str):
  freq_dict = dict()
  for letter in str:
      freq_dict[letter] = freq_dict.get(letter, 0) + 1

  return freq_dict

freq('hello')
```

#### Problem 3

Suppose that Alice and Bob are engaged in key-exchange using the Diffie-Helmann algorithm. See Lecture 9. 

Alice uses a secret number  a
 , and Bob uses a secret number  b.
 You are Eve, and you  know neither  a
  nor b. 
 
 Alice and Bob choose the prime  p=547
  and the fixed number  3
  for their exchange. Alice then sends Bob
3^a % 547=304
 
and Bob sends Alice
3^b % 547=243.
 
This communication is done over a public channel, so you know all the above numbers. 

Both Alice and Bob can then compute the shared secret  3^(ab) % 547
  by exponentiation, using their secret numbers.
You are Eve. Find their shared secret. Hint: find a and b first.

```python
def fast_mod(g, a, p):
  mod = 1
  for i in range(a):
    mod = mod * g % p
  return mod

g = 3
p = 547

#find a
a = 0
while (fast_mod(g, a, p)) != 304:
  a += 1

#find b
b = 0
while (fast_mod(g, b, p)) != 243:
  b += 1

#are these the correct exponents?
if (fast_mod(304, b, p) == fast_mod(243, a, p)):
  #find secret key
  print(fast_mod(g, a*b, p))
```

#### Problem 4

Write a function that takes in three primes p,q,b and generates:

  -a public RSA key (b,n)


  -and private key  e

```python
def rsa_key(p, q, b):
  public_key = (b, p*q)
  private_key = 0

  while (b*private_key) % ((p-1)*(q-1)) != 1:
    private_key += 1

  return public_key, private_key 
```

### Problem 5

Alice is a bank and uses the RSA algorithm to generate a public key (b=7, n=1517). 
Bob uses this public key to encrypt his pin code to send to her, 
and he sends '334'. 
You've interecepted this message. 
What was his pin code?

```python
b = 7
n = 1517

#334 ** e % n
def primes(n):
  isPrime = [True for i in range(n+1)]
  isPrime[0] = False
  isPrime[1] = False
  
  for i in range(2, n+1):
    j = i*2
    while j <= n:
      isPrime[j] = False
      j += i
  
  prime_list = [i for i in range(len(isPrime)) if isPrime[i]]
  return(prime_list)

def factors(n):
  prime_list = primes(n)
  factors = []

  for prime in prime_list:
    while n % prime == 0:
      factors.append(prime)
      n = n/prime

  return(factors)

#find p and q via prime factorization
primes = factors(n)
p = primes[0]
q = primes[1]

#get private key
e = rsa_key(p, q, b)[1]

#decrypt with c**e % pq
decrypted = 334**e % n
print(decrypted)
```
