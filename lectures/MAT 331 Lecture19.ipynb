### A physical dynamical system

Suppose two particles at position x1 and x2 , of mass m1 and m2, have velocities u1 and u2.
If u1>u2 and x1< x2, they will collide at some point.
The new velocities are given by the formula
[here](https://en.wikipedia.org/wiki/Elastic_collision).

1. Suppose a x1=1 and x2=2. Suppose v1=0
and v2=-1.
2. Suppose a particle with velocity v1 hits a wall. After the collision, what is the velocity of the particle?



```python
# 3. Complete the following function.
def velocitiesAfterCollision(m1,m2,v1,v2):
  '''
    Input:
    -- m_i, a positive number representing the mass of particle i
    -- v_i, a number representing the velocity of particle i
    Return:
    -- newVi, the velocities of the particles after they collide
  '''
  newV1 = v1*((m1-m2)/(m1+m2)) + v2*((2*m2)/(m1+m2))
  newV2 = v1*((2*m1)/(m1+m2)) + v2*((m2-m1)/(m1+m2))

  return newV1, newV2
```

```python
print(velocitiesAfterCollision(1, 1, 0, -1))
print(velocitiesAfterCollision(1, 1, 1, 0))
```

```python
# 4. Write a function that simulates one step of the bouncing ball process

def update(m, v1, v2):
  '''
  ball 1 has mass 1
  ball 2 has mass m
  Inputs:
    m -- mass of ball 2
    v1 -- velocity of ball 1
    v2 -- velocity of ball 2
  Returns:
    newV1 -- new velocities after the next collision
    newV2 --
  '''
  #particle 1 hits the wall
  if v1 < 0:
    newV1 = -v1
    newV2 = v2
  #particles collide
  elif v1 > v2:
    newV1, newV2 = velocitiesAfterCollision(1, m, v1, v2)
  #particles never collide
  else:
    newV1 = v1
    newV2 = v2

  return newV1, newV2
```

```python
print(update(1, 1, 2))
```

```python
# 5. Write a function that simulates the bouncing ball process where the 2nd ball
# has mass m.

def simpleSim(m, v1, v2):
  '''Simulates the bouncing ball process:
    Initially, there is a ball at x=1 and another ball at x=2.
    The 2nd ball has mass m.

    Input:
    -- the mass of the second ball
    Returns:
    -- the total number of collisions
  '''
  totalCollisions = 0

  while v1 > v2 or v1 < 0:
    v1, v2 = update(m, v1, v2)
    totalCollisions += 1

  return totalCollisions
```

```python
print(simpleSim(1, 0, -1))
print(simpleSim(100, 0, -1))
print(simpleSim(10000, 0, -1))
print(simpleSim(100000000, 0, -1))
```
