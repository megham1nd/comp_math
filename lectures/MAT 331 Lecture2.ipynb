### Randomness in python

random.random() generates a random real number from 0 to 1.
You can use it to simulate a coin flip.


```python
import random 
random.random()
```

```python
def coin_flip():
  x = random.random()
  print(x)
  if x > 0.5:
    return 'Heads'
  else:
    return 'Tails'
```

#### Problem 1 

Write a function that simulates the position of the man after 10 steps.

```python
def position(n):
  x = 5
  for i in range(n):
    if coin_flip() == "Heads":
      x -= 1
    else:
      x += 1
  return x
```

```python
# Start at x=5
# Flip coin, if heads, x=4, if tails x=6
# ... Keep going for 10 steps
# 

def positionAfterTenSteps():
  x = 5
  for i in range(10):
    result = coin_flip()
    if result == 'Heads':
      x = x-1
    else: 
      x = x+1
  return x 


# Might as well generalize to n steps:
def positionAfterSteps(n):
  x = 5
  for i in range(n):
    result = coin_flip()
    if result == 'Heads':
      x = x-1
    else: 
      x = x+1
  return x 
    
```

### Problem 2 

Write a function that simulates one run of the gamblers ruin process.

```python
def simulation(start_pos):
  x = start_pos
  while x>0 and x<20:
    if coin_flip() == "Heads":
      x -= 1
    else:
      x += 1
  return x
```

```python
# Start x = start
# Flip coin, if tails  x-=1, if heads x+=1
# Keep going until x = 0 or x = 20

def simulation(start):
  x = start
  while x > 0 and x < 20:
    if coin_flip() == 'Heads':
      x += 1
    else: 
      x-= 1
  return x
      
```

Finally, we can answer the motivating question at the beginning of lecture.
We're going to run the experiment 10000 times, and see how many times the man ends up at 0.

```python
def gamblers_probability(n):
  results = [simulation(5) for i in range(n)]
  reaches_zero = [result for result in results if result ==0]
  print(len(reaches_zero)/n)
```

```python
gamblers_probability(10000)
```

```python
# Do the experiment N times:
N = 100000
results = [simulation(1) for i in range(N)]
failures = [result for result in results if result == 0]
len(failures)/N
```
