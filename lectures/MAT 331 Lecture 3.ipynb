Problem 1. Alice and Bob are duelling again. This time, they are using guns.
Alice shoots first, and has probability p of hitting Bob (and winning the duel).
Bob's shots have probability q of hitting Alice (and winning the duel).
Approximate the probability that Alice wins the duel.

```python
import random

def duel(p, q):
  alice = True
  bob = True
  while alice and bob:
    if random.random() < p: #Alice wins
      return 1
    elif random.random() < q: #Bob wins
      return 0

def duel_probability(p, q, N):
  duels = [duel(p, q) for i in range(N)]
  return sum(duels)/N
```

Problem 2. Use the Monte Carlo method to estimate the area of unit disk. Generalize to higher dimensions. 


```python
import random
import math

#disk of radius 1
def area(N):
  hits = 0
  for i in range(N):
    random_x = -1 + 2*random.random()
    random_y = -1 + 2*random.random()
    if (random_x)**2+(random_y)**2 < 1:
        hits += 1
  return hits/N

#generalized for all dimensions using radius 1
def area_dimension(N, d):
  hits = 0
  for i in range(N):
    ori_dist = [random.random()**2 for i in range(d)]
    if sum(ori_dist) < 1:
        hits += 1
  return hits/N
```

Problem 3. Integrate x^2 from 0 to 1 using the Monte Carlo method. 

```python
import random

def integrate(N):
  hits = 0
  for i in range(N):
    random_x = random.random()
    random_y = random.random()
    if random_y < (random_x)^2:
        hits += 1
  return hits/N
```

Problem 4. Birthday paradox. Suppose you invite 26 random people to a party. 
What is the probability that two people have the same birthday?

```python
def birthday_sim():
  birthdays_dict = {i:0 for i  in range(1, 366)}
  for i in range(26):
    birthday = random.randint(1, 365)
    birthdays_dict[birthday] += 1
  for key, value in birthdays_dict.items():
    if value >= 2:
      return 1
  return 0

def birthday_probability(N):
  simulation = [birthday_sim() for i in range(N)]
  return sum(simulation)/N

birthday_probability(1000)
```
