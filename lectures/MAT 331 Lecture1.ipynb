### Loops

The point of using a computer is that you can tell it to do repetitive tasks.

E.g. print the integers from 1 to 100:

```python
for i in range(1,101): # range is [); [inclusive, exclusive)
  print(i)
```

```python
for i in range(10): # assumes start at 0, excludes input integer
  print(i)
```

E.g. find the sum of all integers from 1 to 100 inclusive


```python
total = 0 # resets value of variable; total will otherwise start at wrong value (saves 5050)
for i in range(1,101):
  total += i
print(total)
```

E.g. keep adding 5 until we are above 121

```python
total = 0
while total < 121:
  total += 5
print(total)
```

E.g. compute 2^10 the slow way



```python
total = 1
for i in range(10): # acts as a counter instead of being an index/used variable
  total = 2 * total 
print(total)
```

### Conditionals

E.g. print out the numbers from 1 to 100 that are divisible by 5

```python
for i in range(1,101):
  if i % 5 == 0:
    print(i)
  
```

### Functions

Remember how you can write cos(2) to give the result of cosine applied to the number?

'cos' in as an example of a function. It is a built in function because it comes pre-defined. But you can define your own functions.

Functions allow you to re-use code efficiently.
E.g. write a function that sums the integers from 1 to n.
Evaluate it on n=4 and n=6.

E.g. convert all the previous examples to functions so we can re-use.


```python
# sum numbers from 1 to n
def sum_numbers(n):
  total = 0
  for i in range(1, n+1):
    total += i
  return total
```

```python
sum_numbers(100)
```

```python
sum_numbers(4)
```

```python
sum_numbers(6)
```
