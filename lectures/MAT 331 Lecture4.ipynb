```python
def f(lst):
  seen = set()
  for a in lst:
    if a in seen:
      return True
    seen.add(a)

f([3, 2, 3])
# a = 3
# 3 is not in seen
# seen = {3}
# a = 2
# 2 is not in seen
# seen = {3, 2}
# a = 3
# 3 is in seen
# function breaks out of loop to return True
```

```python
def g(n):
  t = 1
  for i in range(n):
    t = 2*t
  return t

g(3)
# n = 3
# i = 0
# t = 2*1 = 2
# i = 1
# t = 2*2 = 4
# i = 2
# t = 2*4 = 8
# function breaks out of loop
# returns 8
```

```python
import random

def graph_sim(graph, start, endpoints):
  current_pos = start
  while current_pos not in endpoints:
    current_pos = graph[current_pos][random.randint(0, len(graph[current_pos])-1)]
    # print(current_pos)
  return current_pos

def probability_sim(func, N):
  results = [func() for i in range(N)]
  hist_dict = {}
  for result in results:
    if hist_dict.get(result):
      hist_dict[result] += 1
    else:
      hist_dict[result] = 1
  for key, value in hist_dict.items():
    hist_dict[key] = value/N
  return hist_dict

g = [[4, 3, 1], [2, 0], [1,3], [0, 2, 4], [0, 3]]
probability_sim(lambda: graph_sim(g, 3, [0,1]), 10000)

```
