```python
def caeser_encrypt(shift, data):
  alphabet = 'abcdefghijklmnopqrstuvwxyz'
  alpha = {alphabet[i]:i for i in range(26)}
  
  encrypted_string = ''

  for letter in data:
    encrypted_string += alphabet[(alpha[letter] + shift) % 26]
  
  return encrypted_string

caeser_encrypt(3, 'z')

def caeser_decrypt(shift, data):
  return caeser_encrypt(-shift, data)

caeser_decrypt(17, 'xffuafspflivufzexxivrk')
```

```python
def vigenere_encrypt(key, data):
  alphabet = 'abcdefghijklmnopqrstuvwxyz'
  alpha = {alphabet[i]:i for i in range(26)}
  
  encrypted_string = ''
  for index, letter in enumerate(data):
    shift = alpha[key[index%len(key)]]
    encrypted_string += alphabet[(alpha[letter]+shift) % 26]
  
  return encrypted_string

def vigenere_decrypt(key, data):
  alphabet = 'abcdefghijklmnopqrstuvwxyz'
  alpha = {alphabet[i]:i for i in range(26)}

  new_key = ''
  for letter in key:
    new_key += alphabet[(26 - alpha[letter])%26]

  return vigenere_encrypt(new_key, data)
```

```python
vigenere_encrypt('lemon', 'attackatdawn')
```

```python
vigenere_decrypt('lemon', 'lxfopvefrnhr')
```

```python
vigenere_decrypt('lamon', 'lxfopvefrnhr')
```
