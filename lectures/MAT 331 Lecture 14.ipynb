```python
def does_escape(func, initial):
  current = initial
  
  for _ in range(100):
    if abs(current)>2:
      return True
    current = func(current)
  return False
```

```python
does_escape(f, 0+1j)
```

```python
def compute_escaping(func, sub):
  escape_matrix = np.zeros((sub, sub))

  delta = 2/sub

  for i in range(sub):
    for j in range(sub):
      z = complex(-2+delta + i*2*delta, -2+delta + j*2*delta)
      if does_escape(func, z):
        escape_matrix[j, i] = 1
  
  return escape_matrix
```

```python
plt.imshow(compute_escaping(f, 1000))
```
