1. Write a function that lists the primes up to length n, using the procedure outlined in the lecture. 



```python
def primes(n):
  isPrime = [True for i in range(n+1)]
  isPrime[0] = False
  isPrime[1] = False
  
  for i in range(2, n+1):
    j = i*2
    while j <= n:
      isPrime[j] = False
      j += i
  
  prime_list = [i for i in range(len(isPrime)) if isPrime[i]]
  return(prime_list)

primes(11)
```

2. Write a function that returns the prime factors of an integer, sorted from lowest to largest. E.g. 
factors(50)=[2,2,5,5]. 


```python
def factors(n):
  prime_list = primes(n)
  factors = []

  for prime in prime_list:
    while n % prime == 0:
      factors.append(prime)
      n = n/prime

  return(factors)

factors(1256896)
print(factors(270))
print(factors(192))
```

3. Write a function that implements the Euclidean algorithm to find the gcd of two numbers a and b:



```python
def gcd(a, b):
  while a!= 0 and b!= 0:
    temp = b
    b = a % b
    a = temp
  
  if a == 0:
    return b
  else:
    return a

gcd(270, 192)
```

4. Write another function that find the gcd, but this time using the prime factorization function wrote in Q2.



```python
def gcd_2(a, b):
  a_factors = factors(a)
  b_factors = factors(b)

  gcd = 1

  for factor_of_a in a_factors:
    if factor_of_a in b_factors:
      gcd *= factor_of_a
      b_factors.remove(factor_of_a)
  
  return gcd

gcd_2(270, 192)
```
